```python
from sklearn.model_selection import train_test_split
import pandas as pd

```

```python
data = pd.read_csv("..//data//Life_Expectancy_Data.csv",sep=",")
```

```python
# Identify countries with a lot of records
_counts = data['Country'].value_counts()
# Define the threshold for considering a countries to have a lot of records
threshold = 2  
selected_ = _counts[_counts >= threshold].index.tolist()
# Filter the data for selected countries
filtered_data = data[data['Country'].isin(selected_)]
# Split the filtered data into training and testing sets
X_train, X_test = train_test_split(filtered_data, test_size=0.2, random_state=42, stratify=filtered_data['Country'])
#X_test = X_test.drop(columns=['Life expectancy '])
# Saving train and test sets to CSV files
X_train.to_csv('..//data//train.csv', index=False)
X_test.to_csv('..//data//test.csv', index=False)
```
